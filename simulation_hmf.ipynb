```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
sim_num_halos_z5 = np.array([5320211, 1956938, 590716, 172266, 48572, 13050, 3184, 692, 105, 14])
sim_num_halos_z10 = np.array([2395302, 727792, 156725, 29620, 4834, 667, 55, 5, 0, 0])
sim_mass = np.array([0.001, 0.0032, 0.01, 0.0316, 0.1, 0.3162, 1, 3.1623, 10, 31.6228])*1e10
```

```python
plt.plot(sim_mass, sim_num_halos_z5/sim_mass, 'o', label='z=5')
plt.plot(sim_mass, sim_num_halos_z10/sim_mass, 'o', label='z=10')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Mass [$M_\odot$/h]')
plt.ylabel('Number of Halos dN/dlog(M) [$h^3 Mpc^{-3}$]')
plt.legend()
plt.show()
```
